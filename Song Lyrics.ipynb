# Song Lyrics

#### This notebook consists of sentiment polarites. Python module used is TextBlob.
##### TextBlob finds the sentiment of english words and no other languages are included. So polarity of other languages will always be 0.
##### Using the above logic I tried to find out the different languages in which the lyrics of the dataset are written.
##### Interstingly I found HINDI.
##### I used wordcloud with masking to produce a wordcloud 
##### wordcloud contains the words used by the artist "ABBA"

#### calculation of sentiment polarity for finding out the language in which the lyrics are written

```python
polarity_list = []
for i in range(df.shape[0]):
    blob1 = df["text"][i]
    polarity_list.append(TextBlob(blob1).sentiment.polarity)
polarity_list[0]
```

```python
se = pd.Series(polarity_list)
df["sentiment_polarity"]=se.values
```

```python
language_df = df[df["sentiment_polarity"]==0]
language_df
```

```python
# making a language list assuming that sentiments for all english lyrics are found

language_list = []
for i in range(df.shape[0]):
    if(df["sentiment_polarity"][i]==0):
        iso_string = TextBlob(df["text"][i]).detect_language()
        language_list.append(languages.get(alpha2=iso_string).name)
    else:
        language_list.append(languages.get(alpha2="en").name)
```

```python
# added languages column to the dataframe
df["languages"]=pd.Series(language_list).values
```

```python
# lyrics in english, Tagalog,Hindi,French,... used package iso639 from languages (this converts "en" to English locale)
df["languages"].value_counts()
```
